```python
import os
import re
from tqdm import tqdm

def remove_urls(sentence):
    # Function to remove URLs from a sentence
    return re.sub(r'http\S+', '', sentence)

def remove_www(sentence):
    # Function to remove URLs from a sentence
    return re.sub(r'www\S+', '', sentence)

# Input and output folder paths
input_folder = '/content/drive/MyDrive/Colab Notebooks/FNS_2020/training/annual_reports'
output_folder = '/content/drive/MyDrive/Colab Notebooks/FNS_2020/pre_training'

try:
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Iterate over all files in the input folder
    for filename in tqdm(os.listdir(input_folder), desc="Preprocessing Files"):
        input_file_path = os.path.join(input_folder, filename)
        output_file_path = os.path.join(output_folder, f"preprocessed_{filename}")

        # Read and preprocess the content of each file
        with open(input_file_path, 'r') as file:
            file_lines = file.readlines()
            #breaking the lines into sentences
            sentences = [line.strip() for line in file_lines]
            #converting the words into lowercases
            sentences_lower = [sentence.lower() for sentence in sentences]
            #removing url which starts with https
            sentences_lower = [remove_urls(sentence) for sentence in sentences_lower]
            #removing url which starts with www
            sentences_lower = [remove_www(sentence) for sentence in sentences_lower]
            #removing nonalphabetical character
            sentences_lower = [re.sub(r'[^a-zA-Z\s]', '', sentence) for sentence in sentences_lower]
            #joining the sentences
            sentences_lower = [' '.join(sentence.split()) for sentence in sentences_lower if sentence.strip() and len(sentence.split()) > 1]

        # Save the preprocessed text to a new file in the output folder
        with open(output_file_path, 'w') as output_file:
            for sentence_lower in sentences_lower:
                output_file.write(f"{sentence_lower}\n")

        tqdm.write(f"Preprocessed text saved to: {output_file_path}")

except FileNotFoundError:
    print(f"Input folder {input_folder} not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import os
import re

def remove_urls(sentence):
    # Function to remove URLs from a sentence
    return re.sub(r'http\S+', '', sentence)

def remove_www(sentence):
    # Function to remove URLs from a sentence
    return re.sub(r'www\S+', '', sentence)

# Input and output folder paths
input_folder = '/content/drive/MyDrive/Colab Notebooks/FNS_2020/validation/annual_reports'
output_folder = '/content/drive/MyDrive/Colab Notebooks/FNS_2020/pre_val'

try:
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Iterate over all files in the input folder
    for filename in os.listdir(input_folder):
        input_file_path = os.path.join(input_folder, filename)
        output_file_path = os.path.join(output_folder, f"{filename}")

        # Read and preprocess the content of each file
        with open(input_file_path, 'r') as file:
            file_lines = file.readlines()
            sentences = [line.strip() for line in file_lines]
            sentences_lower = [sentence.lower() for sentence in sentences]
            sentences_lower = [remove_urls(sentence) for sentence in sentences_lower]
            sentences_lower = [remove_www(sentence) for sentence in sentences_lower]
            sentences_lower = [re.sub(r'[^a-zA-Z\s]', '', sentence) for sentence in sentences_lower]
            sentences_lower = ["'" + sentence.strip() + "'," for sentence in sentences_lower if sentence.strip() and len(sentence.split()) > 1]

        # Save the preprocessed text to a new file in the output folder
        with open(output_file_path, 'w') as output_file:
            for sentence_lower in sentences_lower:
                output_file.write(f"{sentence_lower}\n")

        print(f"Preprocessed text saved to: {output_file_path}")

except FileNotFoundError:
    print(f"Input folder {input_folder} not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from lexrank import LexRank
from lexrank.mappings.stopwords import STOPWORDS
import pandas as pd
import numpy as np
from path import Path
```

**TRAINING**

```python
documents = []
documents_dir = Path('/content/drive/MyDrive/Colab Notebooks/FNS_2020/pre_training')

for file_path in documents_dir.files('*.txt'):
    with file_path.open(mode='rt', encoding='utf-8') as fp:
        documents.append(fp.readlines())

```

```python
lxr = LexRank(documents, stopwords=STOPWORDS['en'])
```

**Summary For one text file**

```python
file_path = "/content/drive/MyDrive/Colab Notebooks/FNS_2020/pre_val/preprocessed_30778.txt"
with open(file_path, 'r') as file:
    sentences = [line.strip().strip("'") for line in file.readlines()]
```

```python
#This code is for single file summary generation,for testing
# Get summary with classical LexRank algorithm
summary = lxr.get_summary(sentences, summary_size=600, threshold=.9)

# Join the list of sentences into a single string
summary_text = '\n'.join(summary)

# Save the summary to a text file
output_file_path = "/content/drive/MyDrive/Colab Notebooks/FNS_2020/summary_30778.txt"
with open(output_file_path, 'w') as output_file:
    output_file.write(summary_text)

print("Summary saved to:", output_file_path)
```

```python
import sys
from rouge import Rouge

# Increase recursion limit
sys.setrecursionlimit(10000)

# Load the summarized text and gold summary
summary_file_path = "/content/drive/MyDrive/Colab Notebooks/FNS_2020/summary/30778.txt"
gold_summary_file_path = "/content/drive/MyDrive/Colab Notebooks/FNS_2020/validation/gold_summaries/30778_1.txt"

with open(summary_file_path, "r") as summary_file:
    summary_text = summary_file.read()

with open(gold_summary_file_path, "r") as gold_summary_file:
    gold_summary_text = gold_summary_file.read()

# Calculate ROUGE score
rouge = Rouge()
scores = rouge.get_scores(summary_text, gold_summary_text)

print("ROUGE scores:")
print(scores)
```
